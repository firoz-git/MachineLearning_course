
#  Groupby function


1:What was the maximun temprature in each of these 3 cities?
2:what was the average windspeed in each of these 3 cities? 

```python
gp=df.groupby("city")
gp
```

```python
gp.get_group("newyork")
```

```python
for city,city_df in gp:
    print(city)
    print(city_df)
    print("\n")
```
