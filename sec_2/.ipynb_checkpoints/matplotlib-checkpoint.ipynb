# Matplot :Labels, Legend, Grid

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day=[1,2,3,4,5,6]
min_t=[20,26,25,28,27,30]
max_t=[32,34,36,38,37,40]
avrg_t=[30,28,27,30,29,28]


plt.plot(day, min_t) ##day an ivde x. min max avrg ith  3 y aan 
plt.plot(day, max_t)
plt.plot(day, avrg_t)
```

```python
day=[1,2,3,4,5,6]
min_t=[20,26,25,28,27,30]
max_t=[32,34,36,38,37,40]
avrg_t=[30,28,27,30,29,28]


plt.plot(day, min_t) 
plt.plot(day, max_t)
plt.plot(day, avrg_t)

plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Weather")
```

```python
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Weather")


plt.plot(day, min_t, label="min") 
plt.plot(day, max_t, label="max")
plt.plot(day, avrg_t, label="avrg")
plt.legend()  #ith adichale topil min max avrg varoo
```

```python
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.title("Weather")


plt.plot(day, min_t, label="min") 
plt.plot(day, max_t, label="max")
plt.plot(day, avrg_t, label="avrg")
plt.legend(shadow=True,fontsize="large")
plt.grid()
```

# Bar Charts

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
company=["amazon","microsoft","facebook","tesla"]
revenue=[90,100,120,200]
plt.bar(company,revenue )
```

```python
plt.xlabel("revenue in billion usd")
plt.ylabel("company")
plt.title("Stock Value")
plt.bar(company,revenue, label="Revenue")
plt.legend(shadow=True,fontsize="large")

```

```python
profit=[40,50,60,75]
plt.xlabel("revenue in billion usd")
plt.ylabel("company")
plt.title("Stock Value")
plt.bar(company-0.2,revenue, label="Revenue")
plt.bar(company+0.2,profit, label="Profit")
plt.legend(shadow=True,fontsize="large")
```

# Histogram

```python
import matplotlib.pyplot as plt
%matplotlib inline
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar) #ivde bins 10 ennam ayan ullath 
```

# Bins parameter

```python
plt.hist(blood_sugar, bins=4)
```

```python
plt.hist(blood_sugar,rwidth=0.8,bins=4)
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],color="red")
```

# Multiple Data Samples in Histogram

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.9,color=["blue","red"],label=["Men","Women"])
plt.legend(shadow=True)
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],color="red",histtype="step")
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],color="red",orientation="horizontal")
```

# Pie

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values=[1400,600,300,410,250]
exp_labels=["home rent","food","internet","car","others"]
plt.pie(exp_values, labels=exp_labels,shadow=True,autopct="%0.1f%%",radius=1.5)

plt.show()
```

```python
exp_values=[1400,600,300,410,250]
exp_labels=["home rent","food","internet","car","others"]
plt.pie(exp_values, labels=exp_labels,shadow=True,autopct="%0.1f%%",radius=1.5,explode=[0,0,0.3,0,0.2])

plt.show()
plt.savefig("D:/Python course/chart.png",bbox_inches="tight",transparent=True,pad_inches=10)
```
