```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
df=pd.DataFrame(iris.data)
df
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
dfn=df.drop(df[["sepal length (cm)","sepal width (cm)"]],axis="columns")
dfn.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel("length")
plt.ylabel("width")
plt.scatter(dfn["petal length (cm)"],dfn["petal width (cm)"],color="red",marker="D")
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3)
y=model.fit_predict(dfn)
y
```

```python
dfn["cluster"]=y
dfn
```

```python
df1=dfn[dfn.cluster==0]
df1
```

```python
df2=dfn[dfn.cluster==1]
df2
```

```python
df3=dfn[dfn.cluster==2]
df3.head()
```

```python
len(df1),len(df2),len(df3)
```

```python
iris.target_names
```

```python
model.cluster_centers_
```

```python
plt.xlabel("Length")
plt.ylabel("Width")
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"],color="red",marker="D",label="setosa")
plt.scatter(df2["petal length (cm)"],df2["petal width (cm)"],color="blue",marker="D",label="versicolor")
plt.scatter(df3["petal length (cm)"],df3["petal width (cm)"],color="green",marker="D",label="virginica")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color="black",marker="+",label="centroid")
plt.legend(shadow=True,fontsize="large")
```

## Elbow Tech

```python
sse=[]
k_range=range(1,10)
for k in k_range :
    model=KMeans(n_clusters=k)
    model.fit(dfn[["petal length (cm)","petal width (cm)"]])
    sse.append(model.inertia_)
```

```python
sse
```

```python
plt.xlabel("Length")
plt.ylabel("Width")
plt.plot(k_range,sse)
```
